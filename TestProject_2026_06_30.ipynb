# 2. Test từng module

## 2.1. Test Ingestion

```
from document_ai.ingestion.document_loader import DocumentLoader
from document_ai.ingestion.document_profiler import DocumentProfiler
from document_ai.ingestion.page_iterator import PageIterator
from fitz import *
PDF_PATH = "SachBCTHKQKT2021.pdf"

loader = DocumentLoader()
loaded_doc = loader.load(PDF_PATH)

profiler = DocumentProfiler()
document_profile = profiler.process(PDF_PATH)

iterator = PageIterator()
page_records = list(iterator.iter_pages(PDF_PATH))

print("Loaded document:", loaded_doc.to_json(ensure_ascii=False, indent=2))
print("Profile:", document_profile)
print("Page records:", len(page_records))
```

## 2.2 Test Page Extraction

```
from document_ai.pipelines.page_extraction_pipeline import PageExtractionPipeline

page_extraction_pipeline = PageExtractionPipeline()

page_raws = []

for page_record in page_records:
    page_raw = page_extraction_pipeline.process_page(page_record)
    page_raws.append(page_raw)

print("page_raws:", len(page_raws))

for p in page_raws[:5]:
    print("Page:", p.page_number)
    print("Text length:", len(p.raw_text or ""))
    print("Blocks:", len(p.text_blocks or []))
    print("Images:", len(p.images or []))
    print("Drawings:", len(p.drawings or []))
    print("---")
```

## 2.3. Test Page Understanding

```
from document_ai.pipelines.page_understanding_pipeline import PageUnderstandingPipeline

page_understanding_pipeline = PageUnderstandingPipeline()

page_documents = []

for page_raw in page_raws:
    page_doc = page_understanding_pipeline.process(page_raw)
    page_documents.append(page_doc)

print("page_documents:", len(page_documents))

for p in page_documents[:5]:
    print("Page:", p.get("page_number") if isinstance(p, dict) else getattr(p, "page_number", None))
    print("Keys:", p.keys() if isinstance(p, dict) else p.__dict__.keys())
    print("---")
```

## 2.4. Test Document Structure

```
from document_ai.pipelines.document_structure_pipeline import DocumentStructurePipeline

document_structure_pipeline = DocumentStructurePipeline()

document_structure_result = document_structure_pipeline.process(
    page_raws=page_raws
)

print(document_structure_result.keys())

print("Title:", document_structure_result.get("title"))
print("Headings:", len(document_structure_result.get("headings", [])))
print("Sections:", len(document_structure_result.get("sections", [])))
print("Paragraphs:", len(document_structure_result.get("paragraphs", [])))
```

## 2.5. Test Table Understanding

```
from document_ai.pipelines.table_understanding_pipeline import TableUnderstandingPipeline

table_pipeline = TableUnderstandingPipeline()

table_understanding_result = table_pipeline.process(
    page_raws=page_raws
)

print(table_understanding_result.keys())

for key in [
    "table_boundaries",
    "table_grids",
    "table_structures",
    "table_semantics",
    "multi_page_tables",
]:
    print(key, len(table_understanding_result.get(key, [])))
```

## 2.6. Test Cross-page Context

```
from document_ai.pipelines.cross_page_context_pipeline import CrossPageContextPipeline

context_pipeline = CrossPageContextPipeline()

cross_page_context_result = context_pipeline.process(
    page_raws=page_raws
)

print(cross_page_context_result.keys())

for key in [
    "section_links",
    "paragraph_continuations",
    "table_continuations",
    "entities",
    "references",
    "context_graph",
]:
    value = cross_page_context_result.get(key)
    if isinstance(value, list):
        print(key, len(value))
    elif isinstance(value, dict):
        print(key, value.keys())
    else:
        print(key, type(value))
```

## 2.7. Test Knowledge Pipeline

```
from document_ai.pipelines.knowledge_pipeline import KnowledgePipeline

knowledge_pipeline = KnowledgePipeline()
knowledge_result = None
knowledge_result = knowledge_pipeline.process(
    page_raws=page_raws,
    document_structure_result=document_structure_result,
    table_understanding_result=table_understanding_result,
    cross_page_context_result=cross_page_context_result,
)

print(knowledge_result.keys())

for key in [
    "chunks",
    "parent_chunks",
    "child_chunks",
    "table_chunks",
    "evidence",
    "citations",
    "knowledge_graph",
    "metadata_enrichment_result",
]:
    value = knowledge_result.get(key)
    if isinstance(value, list):
        print(key, len(value))
    elif isinstance(value, dict):
        print(key, value.keys())
    else:
        print(key, type(value))
```

## 2.8. Test Indexing

```
from document_ai.indexing.bm25_index_builder import build_bm25_index
from document_ai.indexing.vector_index_builder import build_vector_index
from document_ai.indexing.graph_index_builder import build_graph_index
from document_ai.indexing.metadata_index_builder import build_metadata_index

bm25_index_result = None
bm25_index_result = build_bm25_index(
    page_raws=page_raws,
    knowledge_result=knowledge_result,
    chunk_result=knowledge_result,
    table_chunk_result=knowledge_result,
    evidence_result=knowledge_result,
    metadata_enrichment_result=knowledge_result.get("metadata_enrichment_result", {}),
)

vector_index_result = None
vector_index_result = build_vector_index(
    page_raws=page_raws,
    knowledge_result=knowledge_result,
    chunk_result=knowledge_result,
    table_chunk_result=knowledge_result,
    evidence_result=knowledge_result,
    metadata_enrichment_result=knowledge_result.get("metadata_enrichment_result", {}),
)

graph_index_result = None
graph_index_result = build_graph_index(
    page_raws=page_raws,
    knowledge_result=knowledge_result,
    cross_page_context_result=cross_page_context_result,
    document_structure_result=document_structure_result,
    table_understanding_result=table_understanding_result,
    chunk_result=knowledge_result,
    evidence_result=knowledge_result,
)

metadata_index_result = None
metadata_index_result = build_metadata_index(
    page_raws=page_raws,
    metadata_enrichment_result=knowledge_result.get("metadata_enrichment_result", {}),
    knowledge_result=knowledge_result,
    chunk_result=knowledge_result,
    table_chunk_result=knowledge_result,
    evidence_result=knowledge_result,
    citation_result=knowledge_result,
    document_structure_result=document_structure_result,
    table_understanding_result=table_understanding_result,
    graph_index_result=graph_index_result,
    bm25_index_result=bm25_index_result,
)

print("BM25:", bm25_index_result.get("bm25_index_summary") or bm25_index_result.get("index_summary"))
print("Vector:", vector_index_result.get("vector_index_summary") or vector_index_result.get("index_summary"))
print("Graph:", graph_index_result.get("graph_index_summary") or graph_index_result.get("index_summary"))
print("Metadata:", metadata_index_result.get("metadata_index_summary") or metadata_index_result.get("index_summary"))
```

## 2.9 Test đầy đủ

```
from document_ai.rag.query_classifier import classify_query
from document_ai.rag.query_router import route_query
from document_ai.rag.context_expander import expand_context
from document_ai.rag.evidence_aggregator import aggregate_evidence
from document_ai.rag.prompt_builder import build_prompt
from document_ai.rag.llm_reasoner import reason_with_llm
from document_ai.rag.citation_verifier import verify_citations

query = "Tóm tắt các nội dung chính trong tài liệu và nêu bằng chứng trích dẫn."

classification_result = classify_query(query)

route_result = route_query(
    query=query,
    query_classification_result=classification_result,
    bm25_index_result=bm25_index_result,
    vector_index_result=vector_index_result,
    graph_index_result=graph_index_result,
    metadata_index_result=metadata_index_result,
    table_chunk_result=knowledge_result,
    table_understanding_result=table_understanding_result,
    page_raws=page_raws,
    evidence_result=knowledge_result,
    citation_result=knowledge_result,
    knowledge_result=knowledge_result,
    route_options={
        "use_context_expansion": True,
        "use_evidence_aggregation": True,
        "use_prompt_building": False,
        "use_llm_reasoning": False,
        "use_citation_verification": False,
    },
)

retrieval_result = route_result.get("retrieval_result", {})
expanded_context_result = route_result.get("expanded_context_result", {})
evidence_aggregation_result = route_result.get("evidence_aggregation_result", {})

prompt_result = build_prompt(
    query=query,
    retrieval_result=retrieval_result,
    expanded_context_result=expanded_context_result,
    evidence_aggregation_result=evidence_aggregation_result,
    citation_verification_result={},
    metadata_enrichment_result=knowledge_result.get("metadata_enrichment_result", {}),
    document_profile=document_profile if isinstance(document_profile, dict) else {},
)

reasoner_result = reason_with_llm(
    query=query,
    prompt_result=prompt_result,
    evidence_aggregation_result=evidence_aggregation_result,
    retrieval_result=retrieval_result,
    expanded_context_result=expanded_context_result,
    citation_verification_result={},
)

citation_verification_result = verify_citations(
    page_raws=page_raws,
    answer_text=reasoner_result.get("answer_text", ""),
    citation_result=evidence_aggregation_result,
    evidence_result=evidence_aggregation_result,
    chunk_result=knowledge_result,
    table_understanding_result=table_understanding_result,
    bm25_index_result=bm25_index_result,
    vector_index_result=vector_index_result,
    graph_index_result=graph_index_result,
    metadata_index_result=metadata_index_result,
)

print("Classification:", classification_result.get("query_type"), classification_result.get("query_intent"))
print("Route:", route_result.get("route_name"))
print("Retrieved:", len(retrieval_result.get("retrieved_items", [])))
print("Expanded:", len(expanded_context_result.get("expanded_context_items", [])))
print("Evidence:", len(evidence_aggregation_result.get("aggregated_evidence", [])))
print("Citations:", len(evidence_aggregation_result.get("citations", [])))
print("Answer status:", reasoner_result.get("answer_status"))
print("Confidence:", reasoner_result.get("answer_confidence"))
print("\nANSWER:\n")
print(reasoner_result.get("answer_text", ""))

print("\nCitation verification summary:")
print(citation_verification_result.get("citation_verification_summary", {}))
```
